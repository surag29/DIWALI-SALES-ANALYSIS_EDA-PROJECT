IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

Load the dataset

```python
file_path = "Diwali Sales Data.csv"   # change if needed
df = pd.read_csv(file_path, encoding="unicode_escape")

df.head()
```

Basic data understanding

```python
print("Shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nInfo:")
df.info()

print("\nSummary statistics (numeric):")
df.describe()

print("\nMissing values per column:")
df.isnull().sum()
```

DATA CLEANING

```python

# Drop unnecessary columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)

# Handle missing values
df.dropna(inplace=True)

# Correct data type for Amount
df['Amount'] = df['Amount'].astype(int)

# Check datatypes
df.dtypes
```

EXPLORATORY DATA ANALYSIS

```python
df.columns
```

BY GENDER

```python
gx = sns.countplot(x = 'Gender', data = df)

for bars in gx.containers:
    gx.bar_label(bars)
    
```

```python
df.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gender = df.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y = 'Amount', data = sales_gender)
```

FROM THE ABOVE GRAPH WE CAN SEE THAT MOST OF THE BUYERS ARE FEMALES AND THE PURCHASHING POWER OF FEMALE IS GREATER THEN MALE 

BY AGE GROUP

```python
agex = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')

for bars in agex.containers:
   agex.bar_label(bars)
```

```python
df.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
SALES_AGE = df.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group', y = 'Amount', data = SALES_AGE)
```

FROM THE ABOVE GRAPH WE CAN SAY THAT THE MOST OF THE BUYERS ARE FROM THE  AGE  GROUP BETWEEN 26-35


BY STATE 

```python

sns.set(rc = {'figure.figsize': (25,5)})

sx = sns.countplot(x = 'State', data = df,)


for bars in sx.containers:
   sx.bar_label(bars)
```

```python
sns.set(rc = {'figure.figsize': (25,5)})

SALES_state = df.groupby('State', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(x = 'State', y = 'Orders', data = SALES_state)
```

```python
sns.set(rc = {'figure.figsize': (25,5)})

SALES_state = df.groupby('State', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x = 'State', y = 'Amount', data = SALES_state)
```

FROM THE ABOVE GRAPHS WE CAN SEE THAT MOST OF THE ORDERS AND  TOTALSALES/AMOUNT ARE FROM UTTAR PRADESH MAHARASHTRA ABD KARNATAKA RESPECTIVELY

BY MARITAL STATUS 

```python
sns.set(rc = {'figure.figsize': (10,5)})
mx = sns.countplot(x = 'Marital_Status', data = df,)


for bars in mx.containers:
   mx.bar_label(bars)
```

```python
sns.set(rc = {'figure.figsize': (10,5)})

SALES_status = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x = 'Marital_Status', y = 'Amount', data = SALES_status,hue= 'Gender')
```

FROM THE ABOVE GRAPH WE CAN SAY THAT MOST OF THE BUYERS ARE MARRIED(WOMEN) AND THE HAVE THE HIGH PURCHASING POWER

BY OCCUPATION

```python
sns.set(rc = {'figure.figsize': (20,5)})
ox = sns.countplot(x = 'Occupation', data = df,)


for bars in ox.containers:
   ox.bar_label(bars)
```

```python
sns.set(rc = {'figure.figsize': (20,5)})

SALES_occupation = df.groupby('Occupation', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x = 'Occupation', y = 'Amount', data = SALES_occupation)
```

FROM THE GRAPH WE CAN SAY THAT MOST OF THE BUYERS ARE FROM IT , HEALTHCARE , AVIATION SECTOR

BY PRODUCT CATEGORY

```python
sns.set(rc = {'figure.figsize': (28,5)})
px = sns.countplot(x = 'Product_Category', data = df,)


for bars in px.containers:
   px.bar_label(bars)
```

```python
sns.set(rc = {'figure.figsize': (20,5)})

SALES_occupation = df.groupby('Product_Category', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x = 'Product_Category', y = 'Amount', data = SALES_occupation)
```

FROM THE ABOVE GRAPH WE CAN SAY THAT MOST OF THE SOLD PRODUCTS ARE FROM FOOD , CLOTHING AND APPAREL AND ELECTRONICS AND GADGETS
